```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
```

```python
w00 = torch.tensor(0.5, requires_grad=True)
b00 = torch.tensor(0.1, requires_grad=True)

w01 = torch.tensor(-0.3, requires_grad=True)
b01 = torch.tensor(0.2, requires_grad=True)

w02 = torch.tensor(0.8, requires_grad=True)
b02 = torch.tensor(-0.1, requires_grad=True)
```

```python
z00 = w00 * x + b00
z01 = w01 * x + b01
z02 = w02 * x + b02

a00 = torch.relu(z00)
a01 = torch.relu(z01)
a02 = torch.relu(z02)

print(a00, a01, a02)
```

```python
w10 = torch.tensor(0.4, requires_grad=True)
b10 = torch.tensor(0.0, requires_grad=True)

w11 = torch.tensor(-0.6, requires_grad=True)
b11 = torch.tensor(0.3, requires_grad=True)
```

```python
z10 = w10 * (a00 + a01 + a02) + b10
z11 = w11 * (a00 + a01 + a02) + b11

a10 = torch.sigmoid(z10)
a11 = torch.sigmoid(z11)

print(a10, a11)
```

```python
combined = a10 + a11
tanh_out = torch.tanh(combined)

print(tanh_out)
```

```python
w20 = torch.tensor(1.2, requires_grad=True)
b20 = torch.tensor(-0.4, requires_grad=True)
```

```python
output = w20 * tanh_out + b20
print(output)
```

```python
output.backward()
print(x.grad)
```
